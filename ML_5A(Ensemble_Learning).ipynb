Implement Random Forest Classifier model to predict the safety of the car.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd. read_csv("car_evaluation.csv")
df.head()
```

```python
df_col=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
df.columns=df_col
df.head()
```

```python
selected_features = ['buying', 'maint', 'safety']
X = df[selected_features].copy()
y = df['class']
```

```python
label_encoder = LabelEncoder()
for col in X.columns:
    X[col] = label_encoder.fit_transform(X[col])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python
accuracy = rf_classifier.score(X_test,y_test)*100
print("Accuracy score : ",accuracy)
```

```python
y_pred = rf_classifier.predict(X_test)
```
